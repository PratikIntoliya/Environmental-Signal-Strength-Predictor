```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

# Generate some sample data (replace this with your actual dataset)
np.random.seed(0)
num_samples = 1000
weather = np.random.rand(num_samples, 1)  # Random weather data
location = np.random.rand(num_samples, 1)  # Random location data
signal_strength = weather * 5 + location * 2 + np.random.randn(num_samples, 1)  # Simulated signal strength

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(np.hstack((weather, location)), signal_strength, test_size=0.2, random_state=0)

# Normalize the input features
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Build a simple neural network model
model = keras.Sequential([
    keras.layers.Dense(64, activation='relu', input_shape=(2,)),
    keras.layers.Dense(1)
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train_scaled, y_train, epochs=50, batch_size=32, validation_data=(X_test_scaled, y_test))

# Evaluate the model
loss = model.evaluate(X_test_scaled, y_test)
print(f"Test Loss: {loss}")

# Use the model to make predictions
weather_new = np.array([[0.8]])  # New weather data
location_new = np.array([[0.6]])  # New location data
X_new = scaler.transform(np.hstack((weather_new, location_new)))
signal_strength_pred = model.predict(X_new)
print(f"Predicted Signal Strength: {signal_strength_pred[0][0]}")
```
